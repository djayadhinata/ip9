### Buat kategori berdasarkan lead time: Short, Medium, dan Long

```python
rules = [(data["lead_time"]>90), ((data["lead_time"]>=30)&(data["lead_time"]<90)), (data["lead_time"]<30)]
cat = ["Long","Medium","Short"]
data["L/M/S"] = np.select(rules,cat)
data.head()
```

### Visualisasikan arrival_date_month  untuk melihat bulan dimana pengunjung terbanyak

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y = data['arrival_date_month'].value_counts()
x = data['arrival_date_month'].value_counts().index.tolist()

plt.barh(x,y, color="orange")
plt.title("Arrival Date Month")
plt.xlabel("count")
plt.ylabel("Month")

plt.show()
```
